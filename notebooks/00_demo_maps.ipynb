```python
import folium
from folium.plugins import HeatMap, MarkerCluster
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent  # notebook está em /notebooks
OUT_HTML = PROJECT_ROOT / "reports" / "maps"
OUT_HTML.mkdir(parents=True, exist_ok=True)

# Centro do mapa (média das coords)
center = [gdf["lat"].mean(), gdf["lon"].mean()]

# ===== Mapa 1: Heatmap (Antes) =====
m1 = folium.Map(location=center, zoom_start=12, tiles="CartoDB positron")

before = gdf[gdf["periodo"] == "antes"][["lat", "lon"]].values.tolist()
HeatMap(before, radius=12, blur=14).add_to(m1)

m1.save(OUT_HTML / "folium_heatmap_antes.html")
print("OK:", OUT_HTML / "folium_heatmap_antes.html")


# ===== Mapa 2: MarkerCluster (Antes vs Depois) =====
m2 = folium.Map(location=center, zoom_start=12, tiles="OpenStreetMap")

cluster_before = MarkerCluster(name="antes").add_to(m2)
cluster_after  = MarkerCluster(name="depois").add_to(m2)

for _, row in gdf[gdf["periodo"]=="antes"].sample(600, random_state=1).iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=2,
        tooltip="antes"
    ).add_to(cluster_before)

for _, row in gdf[gdf["periodo"]=="depois"].sample(600, random_state=1).iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=2,
        tooltip="depois"
    ).add_to(cluster_after)

folium.LayerControl().add_to(m2)

m2.save(OUT_HTML / "folium_cluster_antes_depois.html")
print("OK:", OUT_HTML / "folium_cluster_antes_depois.html")
```
